Recursion

```python
def fibonacci(n:int) -> int: 
    if n == 0: 
        return 1 
    if n == 1: 
        return 1
    else: 
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(5)
```

Imagine a tree of computations that then calculate

recursion uses a recursive call stack, this is a stack data structure

calls go on the top of the stack, stack then build up

fibonacci for example: 
fib(5) 

generates the following call stack: 
all of the items in the call stack have addresses for the memory location in the computer

f(4) - f(3) - f(2) - f(1)

f(1) is the base case so does not call

the return goes to the next value in the stack

f(2) gets the return and calls f(0) 

f(2) returns 

f(3) calls f(1)

f(3) returns 

f(4) calls f(2) (goes down its own chain)

f(4) returns


Without memoization multiple calculations are repeated / doubled. 

O(2^N)


Space complexeity is the height of the tree. Typically O(n)

Typical data structures: linked list, graph, tree

Linked List Recursion Implementation.

```python
class SingleLinkedList: 

    def __init__(self, val, next=None): 
        self.val = val
        self.next = next

    def __str__(self): 
        return str(self.val)
```

```python
head = SingleLinkedList(1)
A = SingleLinkedList(2)
B = SingleLinkedList(5)
C = SingleLinkedList(8)

head.next = A 
A.next = B 
B.next = C
C.next = None
```

```python
print(head)
```

```python
# O(n) for time and space, because of the recursive call stack. 
def reverse(node): 
    if not node: 
        return 
    
    reverse(node.next)
    print(node)

reverse(head)
```

Memoization Fibonacci 

```python
class FibonacciAlgos: 
    def __init__(self):
        self.memo = {} 
    
    def fibonacci(self, n:int) -> int: 
        if n == 0: 
            return 1 
        if n == 1: 
            return 1
        else: 
            return self.fibonacci(n-1) + self.fibonacci(n-2)
        
    def mfibonacci(self, n:int):
        if n == 0: 
            return 1 
        if n == 1: 
            return 1
        if n in self.memo: 
            return self.memo[n]
        else: 

            self.memo[n] = self.mfibonacci(n-1) + self.mfibonacci(n-2)
            return self.memo[n]
```

```python
from time import time
start = time()
n = 40
test = FibonacciAlgos().mfibonacci(n)
end = time()
print(test, end - start)
start = time()
n = 40
test = FibonacciAlgos().fibonacci(n)
end = time()
print(test, end - start)
```
